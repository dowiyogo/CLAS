```python
import os

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')

PATH = os.path.abspath("dataplotter.ipynb")
```

```python
matplotlib.rcParams.update({'font.size': 16})
```

```python
def generatefilenames(i, f):
    bgn = "data/ADC_Values_"
    end = ".txt"
    filenames = []
    for i in range(i, f): filenames.append(os.path.join(os.path.dirname(PATH), bgn + str(i) + end))
    return filenames
```

```python
x = []
y = []

for filename in generatefilenames(1910, 1911):
    f = open(filename)
    for l in f.readlines():
        sl = l[:-1].split('\t')
        y.append(int(sl[0]))
```

```python
filename = os.path.join(os.path.dirname(PATH), "data/ADC_Values_1910.txt")
f = open(filename)
```

```python
x = []
y = []

counter = 0
while True:
    counter += 1
    if counter == 10000: break
    l = f.readline()
    sl = l[:-1].split('\t')
    y.append(int(sl[0]))
    x.append(sl[1])

plt.plot(y)
plt.show()
```
